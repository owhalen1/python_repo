```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


# File to Load (Remember to change these)
city_data_to_load = "C:\\Users\\Owner\\Documents\\Data Science Bootcamp\\Homework\\city_data.csv"
ride_data_to_load = "C:\\Users\\Owner\\Documents\\Data Science Bootcamp\\Homework\\ride_data.csv"

# Read the City and Ride Data
city_reader = pd.read_csv(city_data_to_load)
ride_reader = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset by merging data frames
city_ride_df = pd.merge(city_reader, ride_reader, how="left", on="city")


# Display the data table for preview
city_ride_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
#group by city 
# total number of rides per city
rides_per_city = city_ride_df.groupby(['city'])['ride_id'].count()
rides_per_city_df = pd.DataFrame(rides_per_city)
rides_per_city_df
```

```python
#avergae of city fares
city_fares = city_ride_df.groupby(['city'])['fare'].mean()
city_fares_df = pd.DataFrame(city_fares)
city_fares_df
```

```python
#number of drivers per city
drivers_per_city = city_ride_df.groupby(['city'])['driver_count'].unique()
drivers_per_city_df = pd.DataFrame(drivers_per_city)
drivers_per_city_df
```

```python
city_types = city_ride_df.groupby(['city'])['type'].unique()
city_types_df = pd.DataFrame(city_types)
city_types_df
```

```python
city_variables = [city_fares_df, rides_per_city_df, drivers_per_city_df, city_types_df]
city_variable_df = pd.concat(city_variables, sort=False, axis=1)
city_variable_df.rename(columns={'fare':'average_fare ($)','ride_id':'total_rides','driver_count':'total_drivers', 'type':'city_type'}, inplace=True)
city_variable_df.total_drivers = city_variable_df.total_drivers.astype(int) 
city_variable_df

```

```python
#x is total number of rides, y is average fare, bubble color is city type and circle size is driver count
# Build the scatter plots for each city types; use dataframe plot method


colors = ["gold", "lightskyblue", "lightcoral"]
pyber_scatter = city_variable_df.plot.scatter(x= 'total_rides', y= 'average_fare ($)', c=colors, grid=True, figsize=(10,10), title= "Total Number of Rides per City")                                        
pyber_scatter.set(xlabel='Total Rides', ylabel='Average Fare in Dollars')
#plt.legend(['gold', 'lightskyblue', 'lightcoral'],['Urban', 'Suburban', 'Rural'])'
plt.savefig("C:\\Users\\Owner\\Documents\\Data Science Bootcamp\\Homework\\PyberBubbleChart.png")
plt.show()

```

## Total Fares by City Type

```python
#calculate fares percentages
#make fares dataframe
fares_grouped = city_ride_df.groupby(['type'])['fare'].sum()
fares_grouped_df = pd.DataFrame(fares_grouped)
fares_grouped_df
```

```python
#total all of the fares
fare_total = fares_grouped_df['fare'].sum()
print(fare_total)
```

```python
fares_grouped_df['percent'] = fares_grouped_df.loc[:,'fare':].sum(axis=1)/fare_total * 100
fares_grouped_df
```

```python
#create pie chart
colors = ["gold", "lightskyblue", "lightcoral"]
explode = (0.333,0,0)
percents = fares_grouped_df['percent']
fares_py = fares_grouped_df.plot(kind="pie", y='fare', colors=colors, autopct='%1.1f%%', shadow=True, explode=explode, startangle=140, title="Percentage of Fares by City Type")


fares_py.axis("equal")
plt.savefig("C:\\Users\\Owner\\Documents\\Data Science Bootcamp\\Homework\\FaresByCityType.png")
plt.show()
```

# Calculate Total Rides by City Type

```python
#group rides
rides_grouped = city_ride_df.groupby(['type'])['ride_id'].count()
rides_grouped_df = pd.DataFrame(rides_grouped)
rides_grouped_df
    
    
```

```python
# calculate total rides
rides_total = rides_grouped_df['ride_id'].sum()
print(rides_total)
```

# Calculate Total Drivers by City Type 

```python
#calculate total drivers by city type 
drivers_grouped = city_ride_df.groupby(['type'])['driver_count'].sum()
drivers_grouped_df = pd.DataFrame(drivers_grouped)
drivers_grouped_df
    
```
